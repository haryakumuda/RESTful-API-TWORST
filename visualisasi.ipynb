```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot();
```

```python
x = [10,20,30,40,50, 100]
y = np.array(x)**2
plt.plot(x,y)
plt.xlabel("Ini sumbu X")
plt.ylabel("Ini sumbu Y")
plt.title("Ini Tabel coba coba")
plt.show()
```

```python
plt.subplot(131) #1 jumlah baris, #3 jumlah kolom, #1 index
plt.plot(x,y, color = "cyan")
plt.xlabel("Ini sumbu X")
plt.ylabel("Ini sumbu Y")
plt.title("Ini Tabel 1")

plt.subplot(132)
plt.plot(y,x, color = "teal")
plt.xlabel("Ini sumbu X")
plt.ylabel("Ini sumbu Y")
plt.title("Ini Tabel 2")

plt.subplot(133)
plt.plot(x,x, color = "teal")
plt.xlabel("Ini sumbu X")
plt.ylabel("Ini sumbu Y")
plt.title("Ini Tabel 3")

plt.tight_layout()

plt.show()
```

```python
x = np.arange(1,20,0.5)
y = x**2

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("Sumbu X")
axes.set_ylabel("Sumbu Y")
axes.set_title("Judul Grafik")

plt.show()
```

```python
#Plot denga OOP
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.5,0.3,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_xlabel("Ini label X")
axes2.set_ylabel("Ini label Y")
axes1.set_title("ini Judul")
```

```python
fig, axes = plt.subplots(2,1)
axes[0].plot(x,y, color = 'red')
axes[0].set_xlabel("Sumbu X1")
axes[0].set_ylabel("Sumbu Y1")
axes[0].set_title("Judul 1")

axes[1].plot(x,y, color = 'cyan')
axes[1].set_xlabel("Sumbu X2")
axes[1].set_ylabel("Sumbu Y2")
axes[1].set_title("Judul 2")

plt.tight_layout()
```

```python
x = np.arange(1,20,0.5)
y = x**2

fig=plt.figure(figsize=(3,3), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(2,1, figsize=(5,5), dpi=111)
axes[0].plot(x,y)
axes[0].set_title("Grafik1")

axes[1].plot(y,x)
axes[1].set_title("Grafik2")

plt.tight_layout
plt.show()
```
